## 결측데이터 치환

```python
#데이터셋 다시 불러오기
```

```python
df = pd.read_csv("../../data_set/2.데이터 클린징/bicycle.csv")
df.head()
```

```python
#결측치 확인
```

```python
df.isnull().sum()
```

```python
#모든 결측치를 0으로 치환. 자기 자신은 변경되지 않습니다
```

```python
df.fillna(0)
```

```python
#모든 결측치를 이전 값으로 치환
```

```python
df.ffill()
```

```python
#이용거리 결측치 열을 이용거리 평균으로 치환. 자기자신을 변경
```

```python
df['이용거리'] = df['이용거리'].fillna( df['이용거리'].mean() )
df
```

```python
#모든 결측치를 각 열의 평균으로 처리
df.fillna(df.mean(), inplace=True)
df.head()
```

```python
'''
대여거치대      23
반납일시        0
반납대여소번호     0
반납대여소명      0
반납거치대      11
이용시간       11
이용거리       11
'''
df[['대여거치대', '반납거치대', '이용시간', '이용거리']] = df[['대여거치대', '반납거치대', '이용시간', '이용거리']].fillna( df[['대여거치대', '반납거치대', '이용시간', '이용거리']].mean() )
df.head()
```
